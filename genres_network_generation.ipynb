```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = ''
with open('../data.json', 'r') as f:
    df = json.loads(f.read())
print(df[0])
```

get only the genres data

```python
import operator
genres = list(map(operator.itemgetter('genres'), df))
genres[:5]
```

get genres freqs and generate the node db

```python
freqs = {}
for genre in genres:
    for g in genre:
        if not g in freqs:
            freqs[g] = 1
        else:
            freqs[g] += 1
freqs
```

```python
genres_json = []
max_freq = max(freqs.values())
for key, val in freqs.items():
    if val/max_freq > .1:
        genres_json.append({'name':key, 'size':val})
genres_json[:5]
```

```python
len(genres_json)
```

```python
with open('nodes_pruned.json', 'w') as outfile:
    json.dump(genres_json, outfile)
```

computing and generating links between genres

```python
unique = list(freqs.keys())
unique[:10]
```

```python
links = []
seen = [] # to not duplicate links
for genre in unique:
    seen.append(genre)
    for group in genres:
        if genre in group:
            for g in group:
                if not g in seen:
                    links.append([genre, g])
links[:100]
```

```python
from collections import Counter
links_freqs = dict(Counter(map(tuple, links)))
```

```python
links_freqs
```

```python
max_linkage = max(links_freqs.values())
```

```python
final_links = []
for key, val in links_freqs.items():
    if val/max_linkage > .15:
        final_links.append({'source' : key[0], 'target' : key[1], 'val' : val})
final_links[0:20]
```

```python
len(final_links)
```

```python
with open('links_pruned.json', 'w') as outfile:
    json.dump(final_links, outfile)
```
